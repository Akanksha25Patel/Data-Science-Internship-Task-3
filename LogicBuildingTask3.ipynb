1: Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# 1. Identify the highest performance score
highest_score = max(employees.values())

# 2. Handle ties and identify all top performers
top_performers = [name for name, score in employees.items() if score == highest_score]

# 3. Format and display the results
performers_str = ", ".join(top_performers)    
print(f"Top Performers Eligible for Bonus: {performers_str} (Score: {highest_score})")                  
```

2: Search Query Keyword Analysis


```python
import string

query = "Buy mobile phone buy phone online"

# 1. Convert the input to lowercase
query_lower = query.lower()

# 2. Ignore common punctuation
query_clean = query_lower.translate(str.maketrans('', '', string.punctuation))

# Words ko list me split karna
words = query_clean.split()

# 3. Count the frequency of each keyword
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# 4. Display only keywords searched more than once
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

3: Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Valid readings store karne ke liye list
valid_readings = []

# enumerate() use karke humein index (hour) aur value dono mil jate hain
for hour_index, reading_value in enumerate(sensor_readings):
    if reading_value % 2 == 0:
        valid_readings.append((hour_index, reading_value))

# Result display karna
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

4: Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# 1. Email domains ko count karne ke liye dictionary
domain_counts = {}

for email in emails:
    domain = email.split('@')[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1

total_emails = len(emails)

for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

 5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# 1. Average calculate karna
avg_sales = sum(sales) / len(sales)

# 2. 30% above average threshold nikaalna
threshold = avg_sales * 1.30

# 3. Spike detect karna (Day number start with 1)
for i, s in enumerate(sales):
    if s > threshold:
        print(f"Day {i+1}: {s}")
```

6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_counts = {}

# ID counts store karna
for uid in user_ids:
    id_counts[uid] = id_counts.get(uid, 0) + 1

# Sirf duplicates (count > 1) display karna
for uid, count in id_counts.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
